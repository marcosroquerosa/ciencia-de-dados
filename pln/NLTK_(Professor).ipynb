NLTK

```python
import nltk

nltk.download('punkt') # modelo pré-treinado de tokenização.
nltk.download('punkt_tab') # versão mais recente / estrutura alternativa dos dados do Punkt
# nltk.download('popular') pacote que baixa vários recursos comuns de uma vez. , stopwords, wordnet, averaged_perceptron_tagger
nltk.download('stopwords') # Pacote de palavras de termos sem informação relevante
```

```python
# Tokenizar palavras
from nltk.tokenize import word_tokenize

texto = "Este é um exemplo simples de tokenização."
tokens = word_tokenize(texto, language='portuguese')
print(tokens)
```

```python
# Tokenizar frase
from nltk.tokenize import sent_tokenize

texto = "Esta é a primeira frase. Esta é a segunda frase."
frases = sent_tokenize(texto, language='portuguese')
print(frases)
```

```python
from nltk.tokenize import RegexpTokenizer

# r'\w+'
# r'' (raw string)
# \w Significa: “qualquer caractere de palavra” inclui: letras (a-z, A-Z), números (0-9), underscore _
# + “um ou mais”
tokenizer = RegexpTokenizer(r'\w+')
tokens = tokenizer.tokenize("Olá!!! Isso é muito legal :)")
print(tokens)
```

```python
from nltk.corpus import stopwords

stop_words = set(stopwords.words('portuguese'))

# remover palavras comuns (ex: "de", "a", "o", "e")
filtered = [w for w in tokens if w.lower() not in stop_words]
print(filtered)
```

```python
# Normalização
# Lowercase
tokens = [w.lower() for w in tokens]
print(tokens)

# Uppercase
tokens = [w.upper() for w in tokens]
print(tokens)
```

```python
# Remoção de pontuação
import string

tokens = tokenizer.tokenize("Olá!!! Isso é muito legal :)")

tokens = [w for w in tokens if w not in string.punctuation]
print(tokens)
```

```python
# Frequência de palavras
from nltk.probability import FreqDist

tokens = tokenizer.tokenize("As rodas do onibus, giram, giram, pela cidade!")

fdist = FreqDist(tokens)
print(fdist.most_common(5))
```

```python
import pandas as pd
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
from nltk.corpus import stopwords
import nltk

# Sample DataFrame
df = pd.read_csv('data.csv')
text = " ".join(review for review in df['text_column'])

# Optional: Using NLTK for specialized stopwords
# nltk.download('stopwords')
# stop_words = set(stopwords.words('english'))

# Create WordCloud object
wc = WordCloud(width=800, height=400,
               background_color='white',
               stopwords=STOPWORDS).generate(text)

# Plot
plt.figure(figsize=(10, 5))
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```
